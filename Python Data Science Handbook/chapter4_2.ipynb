```python
from sklearn.datasets import load_digits

digits = load_digits(n_class=6)
```

```python
# 用IsoMap方法将数字投影到二维空间 
from sklearn.manifold import Isomap

iso = Isomap(n_components=2)
projection = iso.fit_transform(digits.data)
```

```python
# 画图
plt.scatter(projection[:, 0], projection[:, 1], lw=0.1,
            c=digits.target, cmap=plt.get_cmap('cubehelix', 6))
plt.colorbar(ticks=range(6), label='digit value')
plt.clim(-0.5, 5.5)
```

```python
# 创建一些正态分布数据
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

# 设置坐标轴和网格
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])  # 空出第一列与最后一行
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)
# main_ax.xaxis.set_ticks([])

# 绘制数据
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)
x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical', color='gray')
x_hist.invert_yaxis()  # 翻转坐标轴
y_hist.hist(y, 40, histtype='stepfilled', orientation='horizontal', color='gray')
y_hist.invert_xaxis()  # 翻转坐标轴
```

```python
births = pd.read_csv('births.csv')
```

```python
quartiles = np.percentile(births['births'], [25, 50, 75])
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = births.query('(births > @mu - 5*@sig) & (births < @mu + 5 * @sig)')
births['day'] = births['day'].astype(int)
births.index = pd.to_datetime(10000 * births.year +
                              100 * births.month +
                              births.day, format='%Y%m%d')
births.head()
```

```python
births_by_date = births.pivot_table('births', [births.index.month, births.index.day])
births_by_date.head()
```

```python
import datetime

births_by_date.index = [datetime.date(2012, month, day) for (month, day) in births_by_date.index]
births_by_date.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(births_by_date.index, births_by_date)
```

```python
import matplotlib as mpl

fig, ax = plt.subplots(figsize=(12, 4))  # type: plt.Figure, plt.Axes
births_by_date.plot(ax=ax)
# 添加标签
style = dict(size=10, color='gray')
ax.text('2012-1-1', 3950, "New Year's Day", **style)
ax.text('2012-1-1', 3950, "New Year's Day", **style)
ax.text('2012-7-4', 4250, "Independence Day", ha='center', **style)
ax.text('2012-9-4', 4850, "Labor Day", ha='center', **style)
ax.text('2012-10-31', 4600, "Halloween", ha='right', **style)
ax.text('2012-11-25', 4450, "Thanksgiving", ha='center', **style)
ax.text('2012-12-25', 3850, "Christmas ", ha='right', **style)

ax.set(title='USA births by day of year (1969-1988)',
       ylabel='average daily births')

ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.DayLocator(bymonthday=15))
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%b'))

ax.annotate('Many people die in winter!', xy=('2012-12-22', 4200), xytext=('2012-10-3', 3950),
            arrowprops=dict(facecolor='black', headwidth=5, width=2,
                            headlength=4),
            horizontalalignment='right', verticalalignment='top');
```

```python
def format_func(value, position):
    # 格式化函数
    N = int(np.round(2*value/np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)
```

```python
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
```

```python
# 灰色背景
ax = plt.axes(fc = 'lightgray')
# 设置网格、坐标轴、刻度
ax.set_axisbelow(True) # 设置坐标轴位于最底端
plt.grid(color='white', linestyle='solid')
for spine in ax.spines.values():
    # 设置边框
    spine.set(visible=True, color='red')
    
ax.xaxis.tick_top() # 设置x轴位于最上方
ax.yaxis.tick_left() # 设置y轴位于最左侧
ax.tick_params(colors=(0.5,0.2,0.3), direction='out') # 设置刻度颜色、方向

for tick in ax.get_xticklabels():
    # 设置x刻度颜色
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    # 设置y刻度
    tick.set_color('blue')
    tick.set_size(10)
    
ax.hist(x, edgecolor = '#E0E0E0', color='#166660')
```

```python
def hist_and_lines():
    np.random.seed(0) 
    fig, ax = plt.subplots(1, 2, figsize=(11, 4)) 
    ax[0].hist(np.random.randn(1000)) 
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```
